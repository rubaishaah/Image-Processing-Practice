# Importing Libraries

```python
import numpy as np 
import pandas as pd 
from glob import glob

import cv2
import matplotlib.pyplot as plt
from skimage import io
from skimage.filters import roberts # edge detection
from skimage.filters import sobel # edge detection
from skimage.filters import try_all_threshold # applying thresholds


plt.style.use('ggplot')
```

# Reading in Images


```python
blood_vessel = glob('/kaggle/input/multichannel-glaucoma-benchmark-dataset/blood-vessel/blood-vessel/*.jpg')
full_fundus = glob('/kaggle/input/multichannel-glaucoma-benchmark-dataset/full-fundus/full-fundus/*.jpg')
optic_cup = glob('/kaggle/input/multichannel-glaucoma-benchmark-dataset/optic-cup/optic-cup/*.jpg')
optic_disc = glob('/kaggle/input/multichannel-glaucoma-benchmark-dataset/optic-disc/optic-disc/*.jpg')
```

# Printing original image from dataset

```python
img = io.imread('/kaggle/input/multichannel-glaucoma-benchmark-dataset/full-fundus/full-fundus/BEH-110.png', as_gray=True)
plt.imshow(img)
```

# 1. RGB seperation

```python
def rgb_sep(original):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].imshow(original[:,:,0], cmap='Reds')
    axs[1].imshow(original[:,:,1], cmap='Greens')
    axs[2].imshow(original[:,:,2], cmap='Blues')
    axs[0].axis('off')
    axs[1].axis('off')
    axs[2].axis('off')
    axs[0].set_title('Red channel')
    axs[1].set_title('Green channel')
    axs[2].set_title('Blue channel')
    plt.show()
    return 0
```

# 2. Edge detection + Applying Grayscale

```python
def edge_detection_r(original):
    edge_roberts = roberts(original)
    plt.imshow(original, cmap = 'gray')
    return 0
```

```python
def edge_detection_s(original):
    edge_sobel = sobel(original)
    plt.imshow(original, cmap = 'gray')
    return 0
```

# 3. Applying Threshold to Image

```python
def thresholding(original):
    fig, ax = try_all_threshold(original, figsize=(10,8), verbose=False )
    return 0
```

# 4. Flipping the Image (horizontally)

```python
def flip(image):
return cv2.flip(image, 1)
```

# Applying techniques on full dataset

```python
for file in blood_vessel:  
    image = cv2,imread(file)
    
    # applying techniques
    RGB = rgb_sep(image)
    roberts = edge_detection_r(image)
    sobel = edge_detection_s(image)
    threshold = thresholding(image)
    flip = flip(image)
```
